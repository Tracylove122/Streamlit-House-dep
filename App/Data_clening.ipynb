```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error,r2_score,mean_absolute_percentage_error,mean_squared_error

```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy="median")
df["total_bedrooms"]=imputer.fit_transform(df[["total_bedrooms"]])
```

```python
df.info()
```

```python
df_dummies = pd.get_dummies(df, columns=["ocean_proximity"]).astype(float)
df_dummies
```

```python
# df=pd.concat(df,df_dummies)
df = pd.concat([df, df_dummies], axis=1)
df.head()
```

```python
df.drop(columns=["ocean_proximity"],inplace=True)
```

```python
df.head()
```

```python
x=df.drop(columns=["median_house_value"])
y=df["median_house_value"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=42)
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
import pickle 
```

```python
with open("scaler.pkl","wb") as f:
    pickle.dump(scaler,f)
```

```python
# models=(RandomForestRegressor(),DecisionTreeRegressor())
# for model in models:
#     model.fit(x_train_scaled,y_train)

models=[
    ("RandomForest", RandomForestRegressor()),
    ("DecisionTree", DecisionTreeRegressor())
]
fitted_model={}
for name,model in models:
    model.fit(x_train_scaled, y_train)
    fitted_model[name]=model
```

```python
fitted_model["RandomForest"]
```

```python
fitted_model["DecisionTree"]
```

```python
prediction_test=fitted_model["RandomForest"].predict(x_test_scaled)
prediction_train=fitted_model["RandomForest"].predict(x_train_scaled)

mae_test=mean_absolute_error(y_test,prediction_test)
mae_train=mean_absolute_error(y_train,prediction_train)

r2_score_test=r2_score(y_test,prediction_test)
r2_score_train=r2_score(y_train,prediction_train)
```

```python
prediction_test=fitted_model["DecisionTree"].predict(x_test_scaled)
prediction_train=fitted_model["DecisionTree"].predict(x_train_scaled)

mae_test_dt=mean_absolute_error(y_test,prediction_test)
mae_train_dt=mean_absolute_error(y_train,prediction_train)

r2_score_test_dt=r2_score(y_test,prediction_test)
r2_score_train_dt=r2_score(y_train,prediction_train)
```

```python
comparising_df=pd.DataFrame(
    {
        'model':["RandomForest","DecisionTree"],
        'MAE Train':[mae_train,mae_train_dt],
        'MAE Test':[mae_test,mae_test_dt],
        'R2 TRAIN':[r2_score_train,r2_score_train_dt],
        'R2 TEST':[r2_score_test,r2_score_test_dt]
    }
)

comparising_df
```
